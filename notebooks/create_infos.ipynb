```python
import mne
import os
import numpy as np
from copy import deepcopy

dataset = 'passive'
raw_path = '/run/media/arthur/11D7-8619/camcan/'
random = 43

data_filepath = os.path.join(
	raw_path,
	'cc700/meg/pipeline/release005/BIDSsep/',
	f'derivatives_{dataset}',
	'aa/AA_movecomp_transdef/aamod_meg_maxfilt_00003/',
)

sub_folder = os.path.join(data_filepath, os.listdir(data_filepath)[random])

for file in os.listdir(sub_folder):
	if file.endswith('.fif'):
		raw = mne.io.read_raw_fif(os.path.join(sub_folder, file))
		raw.resample(500)
		break

for ch_type in ('planar1', 'planar2', 'mag'):
	np.save(f'../camcan_info_{ch_type}.npy', deepcopy(raw).pick([ch_type]).info, allow_pickle=True)
```
